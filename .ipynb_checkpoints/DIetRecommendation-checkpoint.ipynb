```python
def summary(df):
    print(f'data shape: {df.shape}')
    summ = pd.DataFrame(df.dtypes, columns=['Data Type'])
    summ['Missing#'] = df.isnull().sum()
    summ['Missing%'] = (df.isnull().sum())/len(df)
    summ['Dups'] = df.duplicated().sum()
    summ['Uniques'] = df.nunique().values
    summ['Count'] = df.count().values
    desc = pd.DataFrame(df.describe(include='all').transpose())
    summ['Min'] = desc['min'].values
    summ['Max'] = desc['max'].values
    summ['Average'] = desc['mean'].values
    summ['Standard Deviation'] = desc['std'].values
    summ['First Value'] = df.loc[0].values
    summ['Second Value'] = df.loc[1].values
    summ['Third Value'] = df.loc[2].values

    display(summ)

summary(df)
```

```python
categorical_columns = df.select_dtypes(include=['category','object']).columns
numerical_columns = df.select_dtypes(exclude=['category','object']).columns
```

```python
dataset=df.copy()
columns=['Food','Breakfast', 'Lunch', 'Dinner','VegNovVeg', 'Junk Food', 'Calories','Fats','Proteins','Iron','Calcium', 'Sodium', 'Potassium', 'Carbohydrates', 'Fibre', 'VitaminD', 'Sugars']
dataset=dataset[columns]
```

```python
max_Calories=2000
max_daily_fat=90
max_daily_Potassium=4700
max_daily_VitaminD=800
max_daily_Iron=18
max_daily_Calcium=1000
max_daily_Sodium=2300
max_daily_Carbohydrate=325
max_daily_Fibre=38
max_daily_Sugars=2000
max_daily_Protein=56
max_list=[max_Calories,max_daily_VitaminD, max_daily_fat,max_daily_Potassium,max_daily_Sodium,max_daily_Carbohydrate,max_daily_Fibre, max_daily_Iron,max_daily_Sugars,max_daily_Protein]
```

```python
extracted_data=dataset.copy()
for column,maximum in zip(extracted_data.columns[6:15],max_list):
    extracted_data=extracted_data[extracted_data[column]<maximum]
```
